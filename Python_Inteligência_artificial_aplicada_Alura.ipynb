```python
import os
from google.colab import userdata
from groq import Groq

# Acessa a chave da API do Grok que voc√™ salvou como um Secret
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

# Inicializa o cliente Groq sem passar a chave diretamente
client_groq = Groq()

completion = client_groq.chat.completions.create(
    model="openai/gpt-oss-20b",
    messages=[
        {
            "role": "user",
            "content": "Testando conex√£o com a API"
        }
    ],
    temperature=1,
    max_completion_tokens=8192,
    top_p=1,
    reasoning_effort="medium",
    stream=True,
    stop=None,
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end="")
```

# Desafio Alura - Python Intelig√™ncia e artificial aplicada:

<br>
1. Criar um arquivo .txt a partir de uma lista de perguntas vindas de uma lista em Python [].
<br>
2. Ler essa as perguntas desse arquivo .txt e salv√°-las em uma lista Python
<br>
3. Obter respostas de um LLM para cada uma.
<br>
4. Salvar os resultados en um novo arquivo.csv
<br>
5. Ler o arquivo.csv usando Pandas

```python
import os
import pandas as pd
from google.colab import userdata
from groq import Groq

lista_desafio = []

with open("perguntas.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        if linha.strip():
            lista_desafio.append(linha.strip())

os.environ["GROQ_API_KEY"] = userdata.get("GROQ_API_KEY")
client_groq = Groq()

lista_de_dicionarios_de_respostas = []

for pergunta in lista_desafio:
    completion = client_groq.chat.completions.create(
        model="openai/gpt-oss-20b",
        messages=[
            {"role": "user", "content": f"Responda de forma sucinta: {pergunta}"}
        ],
        temperature=0.3,
        max_completion_tokens=500,
        top_p=1,
    )

    resposta = completion.choices[0].message.content.strip()
    print(f"{resposta}\n")

    lista_de_dicionarios_de_respostas.append({
        "pergunta": pergunta,
        "resposta": resposta
    })

df_perguntas_e_respostas = pd.DataFrame(lista_de_dicionarios_de_respostas)
df_perguntas_e_respostas.to_csv("respostas.csv", index=False, encoding="utf-8")

df_perguntas_e_respostas = pd.read_csv("respostas.csv")
df_perguntas_e_respostas
```

# Desafio - LLM + Classificador de satisfa√ß√£o:

1. Carregar um arquivo.csv com feedback de clientes (reviews.csv)
2. Usar um LLM para classificar o sentimento de cada feedback (Positivo, Negativo, Neutro)
3. Adicionar uma nova coluna com essa classifica√ß√£o ao DataFrame.

```python
import pandas as pd
import numpy as np

# Gerando dados fict√≠cios
num_linhas = 10
notas = np.random.randint(0, 6, num_linhas) # Notas de 0 a 5
nomes_produtos = [f"Produto_{i+1}" for i in range(num_linhas)]
tipos_produtos = np.random.choice(["Eletr√¥nico", "Livro", "Roupa", "Alimento", "Brinquedo"], num_linhas)
categorias_produtos = np.random.choice(["Categoria A", "Categoria B", "Categoria C", "Categoria D"], num_linhas)

# Gerando depoimentos mais complexos e variados
depoimentos = []
for i in range(num_linhas):
    if notas[i] >= 4:
        depoimentos.append(f"Produto_{i+1} √© excelente! Adorei a qualidade e o desempenho. Recomendo muito!") # Positivo
    elif notas[i] == 3:
        depoimentos.append(f"O Produto_{i+1} √© razo√°vel. Cumpre o que promete, mas n√£o me surpreendeu. Nada de especial.") # Neutro
    else:
        depoimentos.append(f"N√£o gostei do Produto_{i+1}. A qualidade deixou a desejar e n√£o atendeu √†s minhas expectativas. Decepcionado.") # Negativo

# Criando o DataFrame
df_reviews = pd.DataFrame({
    "Nota do produto": notas,
    "Nome do produto": nomes_produtos,
    "Tipo de produto": tipos_produtos,
    "Categoria do produto": categorias_produtos,
    "Depoimento do cliente": depoimentos
})

# Salvando em um arquivo CSV
df_reviews.to_csv("reviews.csv", index=False)

print("Arquivo 'reviews.csv' criado com sucesso com depoimentos mais variados!")
```
